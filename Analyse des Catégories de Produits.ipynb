```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sélectionner les colonnes de ventes pour les 12 mois
sales_columns = ['sales_month_1', 'sales_month_2', 'sales_month_3', 'sales_month_4', 
                 'sales_month_5', 'sales_month_6', 'sales_month_7', 'sales_month_8', 
                 'sales_month_9', 'sales_month_10', 'sales_month_11', 'sales_month_12']

# Transformer les données pour qu'elles soient en format long (tidy)
long_data = data.melt(
    id_vars=["category"], 
    value_vars=sales_columns, 
    var_name="month", 
    value_name="sales"
)

# Extraire les numéros de mois à partir des noms de colonnes (ex: "sales_month_1" → 1)
long_data["month"] = long_data["month"].str.extract("(\d+)").astype(int)

# Créer une grille de FacetGrid pour afficher les courbes de toutes les catégories
g = sns.FacetGrid(long_data, col="category", col_wrap=4, height=4, sharey=False)

# Tracer les courbes pour chaque catégorie
g.map(sns.lineplot, "month", "sales", marker="o")

# Ajouter des titres et labels
g.set_titles("{col_name}")
g.set_axis_labels("Mois", "Ventes")


# Ajouter une grille pour la lisibilité
g.tight_layout()

# Afficher le graphique
plt.show()
```

Clothing :
Une augmentation remarquable des ventes est observée au cours du 3ᵉ mois et du 8ᵉ mois.
En revanche, les ventes de vêtements sont assez faibles en début d'année et au cours du 5ᵉ mois.

Home and Kitchen :
Les ventes présentent une instabilité marquée par une forte volatilité (augmentations et diminutions) d’un mois à l’autre.
On observe un pic de ventes au 4ᵉ mois, tandis qu’elles sont faibles en début d'année.

Toys :
Les ventes sont particulièrement élevées en juin, juillet et août, avec des pics notables en début et en fin d'année.
Le reste de l'année, les ventes de jouets restent faibles.

Books :
Les ventes de livres montrent une grande instabilité d’un mois à l’autre.
Cependant, une performance remarquable est enregistrée au mois d’octobre, où les ventes dépassent les 520 unités.

Electronics :
Les ventes d'électronique sont notables dès le début de l'année, atteignant 550 unités.
Elles diminuent progressivement jusqu’au 8ᵉ mois, où elles atteignent leur niveau le plus bas. Après cela, elles commencent à augmenter constamment avant de baisser à nouveau en décembre.

Health :
Ce domaine connaît des pics élevés et des baisses notables, avec une volatilité qui semble liée aux saisons.
Les ventes sont particulièrement élevées entre les mois 3 et 5, 6 et 8, ainsi que 10 et 12.

Sport :
Les ventes dans ce secteur sont instables, avec un pic en février.
Elles diminuent ensuite jusqu’au mois de mai, où une augmentation temporaire est observée, avant de baisser à nouveau en juillet.
Une nouvelle hausse est visible entre les mois de septembre et octobre
                    

```python
# Calculer la matrice de corrélation
correlation_matrix = data[['price', 'review_score', 'review_count']].corr()
# Créer la carte thermique
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Ajouter un titre
plt.title("Matrice de corrélation", fontsize=16)
plt.show()
```

À partir de cette matrice de corrélation, on peut remarquer que le prix n’a pas d’influence significative sur 
le score des avis (review score) ni sur le nombre d'avis (review count)

```python
# Calculer la somme des review score et review count par catégorie
summary_by_category = data.groupby('category').agg(
    total_review_score=('review_score', 'sum'),
    total_review_count=('review_count', 'sum')
).reset_index()

# Trier par total_review_count en ordre décroissant
summary_by_category = summary_by_category.sort_values(by='total_review_count', ascending=False)

# Afficher le tableau final
print(summary_by_category)
```

Les catégories Sports et Books bénéficient d’une popularité et d’une interaction considérables. Les clients sont particulièrement attirés par ces deux produits, avec une légère préférence pour les livres.
Il serait pertinent d’apporter des améliorations aux produits de la catégorie Sports et de proposer des offres attractives pour renforcer leur position.

En revanche, la catégorie Home & Kitchen est la moins attrayante pour nos clients. Il serait judicieux de renforcer les campagnes de marketing afin d’augmenter sa visibilité et d’attirer davantage l’attention des clients
