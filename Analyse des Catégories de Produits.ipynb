```python
import seaborn as sns
import matplotlib.pyplot as plt

# Slectionner les colonnes de ventes pour les 12 mois
sales_columns = ['sales_month_1', 'sales_month_2', 'sales_month_3', 'sales_month_4', 
                 'sales_month_5', 'sales_month_6', 'sales_month_7', 'sales_month_8', 
                 'sales_month_9', 'sales_month_10', 'sales_month_11', 'sales_month_12']

# Transformer les donnes pour qu'elles soient en format long (tidy)
long_data = data.melt(
    id_vars=["category"], 
    value_vars=sales_columns, 
    var_name="month", 
    value_name="sales"
)

# Extraire les numros de mois  partir des noms de colonnes (ex: "sales_month_1"  1)
long_data["month"] = long_data["month"].str.extract("(\d+)").astype(int)

# Crer une grille de FacetGrid pour afficher les courbes de toutes les catgories
g = sns.FacetGrid(long_data, col="category", col_wrap=4, height=4, sharey=False)

# Tracer les courbes pour chaque catgorie
g.map(sns.lineplot, "month", "sales", marker="o")

# Ajouter des titres et labels
g.set_titles("{col_name}")
g.set_axis_labels("Mois", "Ventes")


# Ajouter une grille pour la lisibilit
g.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Calculer la matrice de corrlation
correlation_matrix = data[['price', 'review_score', 'review_count']].corr()
# Crer la carte thermique
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Ajouter un titre
plt.title("Matrice de corrlation", fontsize=16)
plt.show()
```

 partir de cette matrice de corrlation, on peut remarquer que le prix na pas dinfluence significative sur 
le score des avis (review score) ni sur le nombre d'avis (review count)

```python
# Calculer la somme des review score et review count par catgorie
summary_by_category = data.groupby('category').agg(
    total_review_score=('review_score', 'sum'),
    total_review_count=('review_count', 'sum')
).reset_index()

# Trier par total_review_count en ordre dcroissant
summary_by_category = summary_by_category.sort_values(by='total_review_count', ascending=False)

# Afficher le tableau final
print(summary_by_category)
```

Les catgories Sports et Books bnficient dune popularit et dune interaction considrables. Les clients sont particulirement attirs par ces deux produits, avec une lgre prfrence pour les livres.
Il serait pertinent dapporter des amliorations aux produits de la catgorie Sports et de proposer des offres attractives pour renforcer leur position.

En revanche, la catgorie Home & Kitchen est la moins attrayante pour nos clients. Il serait judicieux de renforcer les campagnes de marketing afin daugmenter sa visibilit et dattirer davantage lattention des clients
